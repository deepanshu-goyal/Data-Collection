# 1.This code is scraping all the Product Manager Jobs in San Francisco, CA listed on indeed.com.
List of attributes are:

1. Job Title
2. Job Location
3. Job Summary
4. Salary Information (if available)

```python
#Import Libraries required for web scraping
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python
#This code will open the chrome browser and then open the indeed website
#To run this code, Provide your chromedriver.exe path
chromelocation = r"C:\Users\Deepanshu Goyal\.wdm\chromedriver\76.0.3809.126\win32\chromedriver.exe" 
driver = webdriver.Chrome(chromelocation)
driver.get('https://www.indeed.com/q-Product-Manager-l-San-Francisco,-CA-jobs.html')
```

```python
# This code will fetch the Title, Location, Salary & Summary information from the first page
job_details = []
for data in driver.find_elements_by_class_name('jobsearch-SerpJobCard.unifiedRow.row.result.clickcard'):
    job_title = data.find_element_by_class_name('jobtitle.turnstileLink ').text
    job_location = data.find_element_by_class_name('location.accessible-contrast-color-location').text
    job_summary = data.find_element_by_class_name('summary').text
    try:
        job_salary = data.find_element_by_class_name('salaryText').text
    except EC.NoSuchElementException:
        job_salary = 'NA'
    job_details.append({'Title':job_title,'Location':job_location,'Salary':job_salary,'Summary':job_summary})
```

```python
#This code will create dynamic url based on the URL pagination pattern
#Read one page at a time and fetch all the above information and then move to next page
#After each page load, this code will wait for 5 sec for any pop-ups. If there is any pop-up it will close the pop-up and 
#start reading the data

url = 'https://www.indeed.com/jobs?q=Product+Manager&l=San+Francisco%2C+CA&start='
for i in range(10,1000,10):
    new_url = url+str(i)
    driver.get(new_url)
    time.sleep(5)
    try:
        driver.find_element_by_class_name('popover-x.ita-popover-x').click()
        print(new_url)
        for data in driver.find_elements_by_class_name('jobsearch-SerpJobCard.unifiedRow.row.result.clickcard'):
            job_title = data.find_element_by_class_name('jobtitle.turnstileLink ').text
            job_location = data.find_element_by_class_name('location.accessible-contrast-color-location').text
            job_summary = data.find_element_by_class_name('summary').text
            try:
                job_salary = data.find_element_by_class_name('salaryText').text
            except EC.NoSuchElementException:
                job_salary = 'NA'
            job_details.append({'Title':job_title,'Location':job_location,'Salary':job_salary,'Summary':job_summary})
    except EC.NoSuchElementException:
        print(new_url)
        for data in driver.find_elements_by_class_name('jobsearch-SerpJobCard.unifiedRow.row.result.clickcard'):
            job_title = data.find_element_by_class_name('jobtitle.turnstileLink ').text
            job_location = data.find_element_by_class_name('location.accessible-contrast-color-location').text
            job_summary = data.find_element_by_class_name('summary').text
            try:
                job_salary = data.find_element_by_class_name('salaryText').text
            except EC.NoSuchElementException:
                job_salary = 'NA'
            job_details.append({'Title':job_title,'Location':job_location,'Salary':job_salary,'Summary':job_summary})
final_output = pd.DataFrame(job_details)
final_output.to_csv('PMscraping.csv', index=False, columns=['Title','Location','Salary','Summary'])
```

```python
#This will close the web browser
driver.close()
```

# 2.This code is scraping naukri.com and fetching all the information technology Jobs in Hyderabad
List of attirbutes are:
1. Company Name
2. job Title
3. Skill

```python
#Import Libraries required for web scraping
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python
#This code will open the chrome browser and then open the indeed website
#To run this code, Provide your chromedriver.exe path
chromelocation = r"C:\Users\Deepanshu Goyal\.wdm\chromedriver\76.0.3809.126\win32\chromedriver.exe" 
driver = webdriver.Chrome(chromelocation)
driver.get('https://www.naukri.com/information-technology-jobs-in-hyderabad-secunderabad')
```

```python
# This code will fetch the Title, Location, Salary & Summary information from the first page
job_details = []
time.sleep(5)
for data in driver.find_elements_by_xpath('//div[@type="tuple"]'):
    job_title = data.find_element_by_id('jdUrl').text
    company_name = data.find_element_by_css_selector('span.org').text
    try:
        skill = data.find_element_by_css_selector('span.skill').text  
    except EC.NoSuchElementException:
        skill = 'NA'
    job_details.append({'Company Name':company_name,'Title':job_title,'Skill':skill})
driver.find_element_by_class_name('grayBtn').click()
```

```python
# This code is used to fetch data from all the pages by clicking the next button 
url = driver.current_url
while url != 'stop':
    time.sleep(5)
    print(url)
    for data in driver.find_elements_by_xpath('//div[@type="tuple"]'):
        job_title = data.find_element_by_id('jdUrl').text
        company_name = data.find_element_by_css_selector('span.org').text
        try:
            skill = data.find_element_by_css_selector('span.skill').text  
        except EC.NoSuchElementException:
            skill = 'NA'
        job_details.append({'Company Name':company_name,'Title':job_title,'Skill':skill})
    if len(driver.find_elements_by_class_name('grayBtn'))>1:
        driver.find_elements_by_class_name('grayBtn')[1].click()
        url = driver.current_url
    else:
        url='stop'     
        
final_data = pd.DataFrame(job_details, columns=['Company Name', 'Title', 'Skill'])
final_data  
```

```python
#This code witll count number of jobs in a company according to the skills
final_data = final_data.groupby(['Company Name','Skill']).count().sort_values(by='Title', ascending=False)
final_data
```

```python
#This code will create csv file 
final_data.to_csv('hyderabad.csv')
```

```python
# This will close the browser
driver.close()
```

# 3.This code is calling News API to fetch latest(i.e.last 20 days) & popular news on "Artifical Intelligence" in english language from all the sources
List of attributes are:
1. Source ID
2. Source Name
3. Author
4. Title
5. Description
6. Content

```python
#Import libraries required for requesting and parsing APIs
import requests
import pandas as pd
```

```python
# Below code is calling News API URL and fetching popular "Artificial Intelligence" news from all the sources 
# Search articles in english language only
# Developer version will show only 100 records
# Output will be in JSON format

url = ('https://newsapi.org/v2/everything?'
       'q="Artificial Intelligence"&'
#      'from=2019-04-01&' #Developer plan permits to request articles as far as 20 days from current date
       'sortBy=popularity&'
       'pageSize=100&' #  Developer accounts are limited to a max of 100 results
       'language=en&'
       'apiKey=4f56d29e685e44beaca8f92efa288023')
fetch_data = requests.get(url)
article_details = fetch_data.json()
article_details
```

```python
# This code is used to parse the JSON 
# For each article extract the following things â€“[source-id , source-name, author, title, description, content] 
# Create csv file 'output_api.csv'

article=[]
for i in article_details['articles']:
    source_id=''
    source_name=''
    author=''
    title=''
    description=''
    content=''

    for k,v in i.items():
        if isinstance(v,dict):
            for key,value in v.items():
                if key=='id':
                    source_id=v[key]
                else:
                    source_name=v[key]   
        elif k=='author':
            author=i[k]
        elif k=='title':
            title=i[k]
        elif k=='description':
            description=i[k]
        elif k=='content':
            content = i[k]
    article.append({'Source_ID':source_id, 'Source_Name':source_name,'Author':author,
                            'Title':title,'Description':description,'Content':content}
                   )

    final_data = pd.DataFrame(article,columns=['Source_ID','Source_Name','Author','Title','Description','Content'])
final_data.to_csv('news_api.csv',index=False)
final_data
```
