```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd
import numpy as np
```

```python
locationOfWebdriver = r"C:\Users\Deepanshu Goyal\.wdm\chromedriver\76.0.3809.126\win32\chromedriver.exe"
driver = webdriver.Chrome(locationOfWebdriver)
```

```python
url = "https://www.edustoke.com/search/day-school/gurugram--MjguNDU5NDk3LDc3LjAyNjYzOA==/?list=city&rangemax=10&rangemin=0"
driver.get(url)
school_name=[]
school_link=[]
school_rating=[]
school_address=[]
school_info=[]

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

time.sleep(100)

for name in driver.find_elements_by_class_name('school_link'):
    school_name.append(name.text)
    school_link.append(name.get_attribute('href'))

for address in driver.find_elements_by_class_name('address_text'):
    school_address.append(address.text)

for info in driver.find_elements_by_css_selector('ul.list-info.search_page'):
    school_info.append(info.text)

data = {'School Name':school_name,'School Link':school_link, 'School Address':school_address,'School Information': school_info}
final_data = pd.DataFrame(data)
final_data
```

```python
final_data.to_csv("School_data.csv")
```
