# This code is calling News API to fetch latest(i.e.last 20 days) & popular news on "Artifical Intelligence" in english language from all the sources. 
List of attributes are:
1. Source ID
2. Source Name
3. Author
4. Title
5. Description
6. Content

```python
#Import libraries required for requesting and parsing APIs
import requests
import pandas as pd
```

```python
# Below code is calling News API URL and fetching popular "Artificial Intelligence" news from all the sources 
# Search articles in english language only
# Developer version will show only 100 records
# Output will be in JSON format

url = ('https://newsapi.org/v2/everything?'
       'q="Artificial Intelligence"&'
#      'from=2019-04-01&' #Developer plan permits to request articles as far as 20 days from current date
       'sortBy=popularity&'
       'pageSize=100&' #  Developer accounts are limited to a max of 100 results
       'language=en&'
       'apiKey=4f56d29e685e44beaca8f92efa288023')
fetch_data = requests.get(url)
article_details = fetch_data.json()
article_details
```

```python
# This code is used to parse the JSON 
# For each article extract the following things –[source-id , source-name, author, title, description, content] 
# Create csv file 'output_api.csv'

article=[]
for i in article_details['articles']:
    source_id=''
    source_name=''
    author=''
    title=''
    description=''
    content=''

    for k,v in i.items():
        if isinstance(v,dict):
            for key,value in v.items():
                if key=='id':
                    source_id=v[key]
                else:
                    source_name=v[key]   
        elif k=='author':
            author=i[k]
        elif k=='title':
            title=i[k]
        elif k=='description':
            description=i[k]
        elif k=='content':
            content = i[k]
    article.append({'Source_ID':source_id, 'Source_Name':source_name,'Author':author,
                            'Title':title,'Description':description,'Content':content}
                   )

    final_data = pd.DataFrame(article,columns=['Source_ID','Source_Name','Author','Title','Description','Content'])
final_data.to_csv('news_api.csv',index=False)
final_data
```
